```python
import pandas as pd

# Load your dataset
df = pd.read_csv("merged_final.csv")

# Standardize columns
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Detect date column
date_cols = [c for c in df.columns if "date" in c]
if not date_cols:
    raise ValueError("No date column found.")
date_col = date_cols[0]

print("Using date column:", date_col)

# Parse the date safely using a strict format
df["date_parsed"] = pd.to_datetime(
    df[date_col],
    format="%d/%m/%y",    # handles D/MM/yy or DD/MM/yy
    errors="coerce"       # invalid formats → NaT
)

# Extract components
df["year"] = df["date_parsed"].dt.year
df["month"] = df["date_parsed"].dt.month
df["day"] = df["date_parsed"].dt.day
df["day_of_year"] = df["date_parsed"].dt.dayofyear

# Check for parsing issues
missing = df[df["date_parsed"].isna()]
print("\nTotal rows:", len(df))
print("Invalid or unrecognized dates:", len(missing))

if len(missing) > 0:
    print("\nExample wrong rows:")
    print(missing[[date_col]].head(10))

# Save final dataset
df.to_csv("merged_final_with_date_features.csv", index=False)
print("\nSaved full dataset with date features → merged_final_with_date_features.csv")

# Save only extracted date fields (optional)
date_extract_df = df[[date_col, "date_parsed", "year", "month", "day", "day_of_year"]]
date_extract_df.to_csv("date_extracted_columns.csv", index=False)
print("Saved date-only file → date_extracted_columns.csv")
```
