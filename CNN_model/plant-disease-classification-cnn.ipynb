```python
# ============================================================
#                IMPORTS
# ============================================================
import os
import shutil
import numpy as np
from pathlib import Path
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader

from torchvision import datasets, transforms, models
from PIL import Image, ImageFile

import matplotlib.pyplot as plt

# Fix corrupted images + PNG transparency issues
ImageFile.LOAD_TRUNCATED_IMAGES = True



# ============================================================
#      Safe loader for corrupted images
# ============================================================
def safe_loader(path):
    img = Image.open(path)
    return img.convert("RGB")



# ============================================================
#                DEVICE
# ============================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)



# ============================================================
#                DATASET SPLITTING FUNCTION
# ============================================================
def split_and_save_dataset(
    source_dir,
    output_dir='/kaggle/working/split_data',
    train_ratio=0.7,
    val_ratio=0.1,
    test_ratio=0.2,
    seed=42
):
    np.random.seed(seed)
    source_path = Path(source_dir)
    output_path = Path(output_dir)

    assert abs(train_ratio + val_ratio + test_ratio - 1.0) < 1e-6, \
        "Ratios must sum to 1.0"

    for split in ['train', 'val', 'test']:
        (output_path / split).mkdir(parents=True, exist_ok=True)

    class_dirs = [d for d in source_path.iterdir() if d.is_dir()]
    if not class_dirs:
        raise ValueError(f"No class folders found in {source_dir}")

    print(f"\nSplitting dataset from '{source_dir}' to '{output_dir}'")
    print(f"Ratios: Train {train_ratio:.0%}, Val {val_ratio:.0%}, Test {test_ratio:.0%}")
    print(f"{'Class':<25}{'Total':<10}{'Train':<10}{'Val':<10}{'Test':<10}")
    print("-" * 60)

    total_train = total_val = total_test = 0
    exts = ['.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff']

    for class_dir in tqdm(class_dirs, desc="Processing classes"):
        class_name = class_dir.name

        # create class folders
        for split in ['train', 'val', 'test']:
            (output_path / split / class_name).mkdir(exist_ok=True)

        files = [f for ext in exts for f in class_dir.glob(f'*{ext}') if f.is_file()]
        if not files:
            continue

        np.random.shuffle(files)
        n_total = len(files)
        n_train = int(n_total * train_ratio)
        n_val = int(n_total * val_ratio)
        n_test = n_total - n_train - n_val

        splits = {
            'train': files[:n_train],
            'val': files[n_train:n_train + n_val],
            'test': files[n_train + n_val:]
        }

        for split, split_files in splits.items():
            for f in split_files:
                shutil.copy2(f, output_path / split / class_name / f.name)

        total_train += n_train
        total_val += n_val
        total_test += n_test

        print(f"{class_name:<25}{n_total:<10}{n_train:<10}{n_val:<10}{n_test:<10}")

    print("\nDataset split completed.")
    print(f"Train: {total_train}, Val: {total_val}, Test: {total_test}")
    print(f"Total images: {total_train + total_val + total_test}")



# ============================================================
#  SPLIT ORIGINAL DATASET (only run once)
# ============================================================
split_and_save_dataset(
    source_dir="/kaggle/input/20k-multi-class-crop-disease-images/Train",
    output_dir="/kaggle/working/split_data",
    train_ratio=0.7,
    val_ratio=0.1,
    test_ratio=0.2,
    seed=42
)



# ============================================================
# Remove bad/corrupted class from all splits
# ============================================================
base = Path('/kaggle/working/split_data')
class_name = 'bollrot on Cotton'  # Change if needed

for split in ['train', 'val', 'test']:
    p = base / split / class_name
    if p.exists():
        shutil.rmtree(p)
        print(f"Deleted {p}")



# ============================================================
#                FINAL PATHS
# ============================================================
train_path = "/kaggle/working/split_data/train/"
valid_path = "/kaggle/working/split_data/val/"
test_path  = "/kaggle/working/split_data/test/"



# ============================================================
#                TRANSFORMS
# ============================================================
train_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.RandomResizedCrop(224, scale=(0.9, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406], [0.229,0.224,0.225])
])

valid_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406], [0.229,0.224,0.225])
])



# ============================================================
#                DATASETS + DATALOADERS
# ============================================================
train_dataset = datasets.ImageFolder(train_path, transform=train_transform, loader=safe_loader)
valid_dataset = datasets.ImageFolder(valid_path, transform=valid_transform, loader=safe_loader)

batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=4)
valid_loader = DataLoader(valid_dataset, batch_size=batch_size, shuffle=False, num_workers=4)

num_classes = len(train_dataset.classes)
class_names = train_dataset.classes

print(f"Detected {num_classes} classes.")



# ============================================================
#          PLOT CLASS DISTRIBUTION
# ============================================================
def plot_class_distribution(dataset):
    counts = np.bincount(dataset.targets)
    plt.figure(figsize=(14,6))
    plt.bar(class_names, counts)
    plt.xticks(rotation=90)
    plt.title("Class Distribution")
    plt.tight_layout()
    plt.show()

plot_class_distribution(train_dataset)



# ============================================================
#                CNN MODEL
# ============================================================
class ImprovedCNN(nn.Module):
    def __init__(self, num_classes):
        super(ImprovedCNN, self).__init__()
        
        self.conv1 = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1), nn.BatchNorm2d(64), nn.ReLU(),
            nn.Conv2d(64, 64, 3, padding=1), nn.BatchNorm2d(64), nn.ReLU(),
            nn.MaxPool2d(2,2), nn.Dropout2d(0.1)
        )

        self.conv2 = nn.Sequential(
            nn.Conv2d(64, 128, 3, padding=1), nn.BatchNorm2d(128), nn.ReLU(),
            nn.Conv2d(128, 128, 3, padding=1), nn.BatchNorm2d(128), nn.ReLU(),
            nn.MaxPool2d(2,2), nn.Dropout2d(0.2)
        )

        self.conv3 = nn.Sequential(
            nn.Conv2d(128, 256, 3, padding=1), nn.BatchNorm2d(256), nn.ReLU(),
            nn.Conv2d(256, 256, 3, padding=1), nn.BatchNorm2d(256), nn.ReLU(),
            nn.MaxPool2d(2,2), nn.Dropout2d(0.3)
        )

        self.conv4 = nn.Sequential(
            nn.Conv2d(256, 512, 3, padding=1), nn.BatchNorm2d(512), nn.ReLU(),
            nn.Conv2d(512, 512, 3, padding=1), nn.BatchNorm2d(512), nn.ReLU(),
            nn.MaxPool2d(2,2), nn.Dropout2d(0.4)
        )

        self.gap = nn.AdaptiveAvgPool2d((1,1))

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(512, 512), nn.BatchNorm1d(512), nn.ReLU(), nn.Dropout(0.5),
            nn.Linear(512, 256), nn.ReLU(), nn.Dropout(0.5),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.conv4(x)
        x = self.gap(x)
        x = self.classifier(x)
        return x



# ============================================================
#                MODEL PATHS
# ============================================================
cnn_weights_path    = "/kaggle/working/CNN_Plant_best.pth"
resnet_weights_path = "/kaggle/working/ResNet_Plant_best.pth"



# ============================================================
#                INITIALIZE CNN
# ============================================================
cnn_model = ImprovedCNN(num_classes).to(device)

if os.path.exists(cnn_weights_path):
    print("Loading CNN saved weights...")
    cnn_model.load_state_dict(torch.load(cnn_weights_path, map_location=device))
else:
    print("No CNN checkpoint found — training from scratch.")



# ============================================================
#                INITIALIZE RESNET50
# ============================================================
if os.path.exists(resnet_weights_path):
    print("Loading ResNet50 saved weights...")
    resnet_model = models.resnet50(weights=None)
    resnet_model.fc = nn.Linear(resnet_model.fc.in_features, num_classes)
    resnet_model.load_state_dict(torch.load(resnet_weights_path, map_location=device))

else:
    try:
        print("Loading pretrained ResNet50 weights...")
        resnet_model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)
        resnet_model.fc = nn.Linear(resnet_model.fc.in_features, num_classes)
    except:
        print("Pretrained weights unavailable — training from scratch.")
        resnet_model = models.resnet50(weights=None)
        resnet_model.fc = nn.Linear(resnet_model.fc.in_features, num_classes)

resnet_model = resnet_model.to(device)



# ============================================================
#                TRAINING LOOP
# ============================================================
def train_model(model, train_loader, valid_loader, num_epochs=50, save_path=None, patience=5):

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=0.001)
    best_val_loss = float("inf")
    epochs_no_improve = 0

    history = {"train_loss":[], "train_acc":[], "val_loss":[], "val_acc":[]}

    for epoch in range(num_epochs):

        # ---------------- TRAIN ----------------
        model.train()
        running_loss = 0
        correct = 0
        total = 0
        
        for imgs, labels in train_loader:
            imgs, labels = imgs.to(device), labels.to(device)
            
            optimizer.zero_grad()
            outputs = model(imgs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, preds = outputs.max(1)
            total += labels.size(0)
            correct += preds.eq(labels).sum().item()

        train_loss = running_loss / len(train_loader)
        train_acc = 100 * correct / total

        # ---------------- VALIDATE ----------------
        model.eval()
        correct = 0
        total = 0
        val_loss = 0

        with torch.no_grad():
            for imgs, labels in valid_loader:
                imgs, labels = imgs.to(device), labels.to(device)
                outputs = model(imgs)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                _, preds = outputs.max(1)
                total += labels.size(0)
                correct += preds.eq(labels).sum().item()

        val_loss /= len(valid_loader)
        val_acc = 100 * correct / total

        # ---------------- HISTORY ----------------
        history["train_loss"].append(train_loss)
        history["train_acc"].append(train_acc)
        history["val_loss"].append(val_loss)
        history["val_acc"].append(val_acc)

        print(f"Epoch {epoch+1}/{num_epochs} | "
              f"Train Acc {train_acc:.2f}% | Val Acc {val_acc:.2f}% | Val Loss {val_loss:.4f}")

        # ---------------- CHECKPOINT ----------------
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            epochs_no_improve = 0
            if save_path:
                torch.save(model.state_dict(), save_path)
                print(f"Checkpoint saved: {save_path}")
        else:
            epochs_no_improve += 1

        # ---------------- EARLY STOPPING ----------------
        if epochs_no_improve >= patience:
            print(f"EARLY STOPPING activated after {epoch+1} epochs.")
            break

    return history



# ============================================================
#                PLOT TRAINING HISTORY
# ============================================================
def plot_history(history, name="Model"):
    plt.figure(figsize=(12,5))

    plt.subplot(1,2,1)
    plt.plot(history['train_loss'],'b-',label='Train Loss')
    plt.plot(history['val_loss'],'r-',label='Val Loss')
    plt.title(f"{name} Loss")
    plt.legend()

    plt.subplot(1,2,2)
    plt.plot(history['train_acc'],'b-',label='Train Acc')
    plt.plot(history['val_acc'],'r-',label='Val Acc')
    plt.title(f"{name} Accuracy")
    plt.legend()

    plt.show()



# ============================================================
#                TRAIN MODELS
# ============================================================
history_cnn = train_model(
    cnn_model, train_loader, valid_loader,
    num_epochs=50, save_path=cnn_weights_path, patience=7
)

plot_history(history_cnn, "CNN_Model")


history_resnet = train_model(
    resnet_model, train_loader, valid_loader,
    num_epochs=50, save_path=resnet_weights_path, patience=7
)

plot_history(history_resnet, "ResNet50_Model")



# ============================================================
#                SINGLE IMAGE PREDICTION
# ============================================================
def predict_single_image(model, img_path, class_names):
    model.eval()

    img = Image.open(img_path).convert("RGB")
    transform = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485,0.456,0.406], [0.229,0.224,0.225]),
    ])

    img_tensor = transform(img).unsqueeze(0).to(device)

    with torch.no_grad():
        output = model(img_tensor)
        probs = torch.softmax(output, dim=1)
        confidence, pred_idx = torch.max(probs, dim=1)

    predicted_class = class_names[pred_idx.item()]
    confidence_score = confidence.item() * 100

    plt.imshow(img)
    plt.axis("off")
    plt.title(f"{predicted_class} ({confidence_score:.2f}%)")
    plt.show()

    return predicted_class, confidence_score
```

```python
# ============================================================
#                IMPORTS
# ============================================================
import os
import shutil
import numpy as np
from pathlib import Path
from tqdm import tqdm
import pandas as pd

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader

from torchvision import datasets, transforms, models
from PIL import Image, ImageFile

import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score

ImageFile.LOAD_TRUNCATED_IMAGES = True  # Fix truncated images


# ============================================================
#                SAFE IMAGE LOADER
# ============================================================
def safe_loader(path):
    img = Image.open(path)
    return img.convert("RGB")


# ============================================================
#                DEVICE
# ============================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)


# ============================================================
#          SPLIT DATASET INTO TRAIN/VAL/TEST (if needed)
# ============================================================
# If you've already run your split and have /kaggle/working/split_data, skip this.
# Otherwise uncomment and adapt source_dir then run.
#
# def split_and_save_dataset(...):
#     ...
#
# split_and_save_dataset(source_dir="/kaggle/input/20k-multi-class-crop-disease-images/Train",
#                        output_dir="/kaggle/working/split_data")

# ============================================================
#                DATASET PATHS (adjust if necessary)
# ============================================================
train_path = "/kaggle/working/split_data/train/"
valid_path = "/kaggle/working/split_data/val/"
test_path  = "/kaggle/working/split_data/test/"

assert os.path.exists(train_path), f"Train path not found: {train_path}"
assert os.path.exists(valid_path), f"Val path not found: {valid_path}"
assert os.path.exists(test_path),  f"Test path not found: {test_path}"


# ============================================================
#                TRANSFORMS
# ============================================================
train_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.RandomResizedCrop(224, scale=(0.9, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406], [0.229,0.224,0.225])
])

valid_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406], [0.229,0.224,0.225])
])

test_transform = valid_transform


# ============================================================
#                DATASETS / DATALOADERS
# ============================================================
train_dataset = datasets.ImageFolder(train_path, transform=train_transform, loader=safe_loader)
valid_dataset = datasets.ImageFolder(valid_path, transform=valid_transform, loader=safe_loader)
test_dataset  = datasets.ImageFolder(test_path,  transform=test_transform,  loader=safe_loader)

batch_size = 32
num_workers = 4

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=num_workers)
valid_loader = DataLoader(valid_dataset, batch_size=batch_size, shuffle=False, num_workers=num_workers)
test_loader  = DataLoader(test_dataset,  batch_size=batch_size, shuffle=False, num_workers=num_workers)

num_classes = len(train_dataset.classes)
class_names = train_dataset.classes

print(f"Detected {num_classes} classes: {class_names[:10]}{'...' if len(class_names)>10 else ''}")


# ============================================================
#         PLOT CLASS DISTRIBUTION (optional)
# ============================================================
def plot_class_distribution(dataset, title="Class distribution"):
    counts = np.bincount(dataset.targets)
    plt.figure(figsize=(14,6))
    plt.bar(range(len(counts)), counts)
    plt.xticks(range(len(counts)), dataset.classes, rotation=90)
    plt.title(title)
    plt.tight_layout()
    plt.show()

# plot_class_distribution(train_dataset)


# ============================================================
#                CNN MODEL
# ============================================================
class ImprovedCNN(nn.Module):
    def __init__(self, num_classes):
        super().__init__()

        self.conv1 = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1), nn.BatchNorm2d(64), nn.ReLU(),
            nn.Conv2d(64, 64, 3, padding=1), nn.BatchNorm2d(64), nn.ReLU(),
            nn.MaxPool2d(2), nn.Dropout2d(0.1)
        )

        self.conv2 = nn.Sequential(
            nn.Conv2d(64, 128, 3, padding=1), nn.BatchNorm2d(128), nn.ReLU(),
            nn.Conv2d(128, 128, 3, padding=1), nn.BatchNorm2d(128), nn.ReLU(),
            nn.MaxPool2d(2), nn.Dropout2d(0.2)
        )

        self.conv3 = nn.Sequential(
            nn.Conv2d(128, 256, 3, padding=1), nn.BatchNorm2d(256), nn.ReLU(),
            nn.Conv2d(256, 256, 3, padding=1), nn.BatchNorm2d(256), nn.ReLU(),
            nn.MaxPool2d(2), nn.Dropout2d(0.3)
        )

        self.conv4 = nn.Sequential(
            nn.Conv2d(256, 512, 3, padding=1), nn.BatchNorm2d(512), nn.ReLU(),
            nn.Conv2d(512, 512, 3, padding=1), nn.BatchNorm2d(512), nn.ReLU(),
            nn.MaxPool2d(2), nn.Dropout2d(0.4)
        )

        self.gap = nn.AdaptiveAvgPool2d((1, 1))

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(512, 512), nn.BatchNorm1d(512), nn.ReLU(), nn.Dropout(0.5),
            nn.Linear(512, 256), nn.ReLU(), nn.Dropout(0.5),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.conv4(x)
        x = self.gap(x)
        return self.classifier(x)


# ============================================================
#                MODEL PATHS
# ============================================================
cnn_weights    = "/kaggle/working/CNN_Plant_best.pth"
resnet_weights = "/kaggle/working/ResNet_Plant_best.pth"
vgg_weights    = "/kaggle/working/VGG19BN_Plant_best.pth"


# ============================================================
#                INITIALIZE MODELS
# ============================================================
# CNN
cnn_model = ImprovedCNN(num_classes).to(device)
if os.path.exists(cnn_weights):
    print("Loading CNN checkpoint...")
    cnn_model.load_state_dict(torch.load(cnn_weights, map_location=device))
else:
    print("CNN: no checkpoint found - training from scratch.")

# ResNet50 (pretrained if available)
try:
    resnet_model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)
    print("ResNet50: loaded pretrained weights.")
except Exception as e:
    print("ResNet50: pretrained weights unavailable, initializing randomly.", e)
    resnet_model = models.resnet50(weights=None)

resnet_model.fc = nn.Linear(resnet_model.fc.in_features, num_classes)
resnet_model = resnet_model.to(device)
if os.path.exists(resnet_weights):
    print("Loading ResNet checkpoint...")
    resnet_model.load_state_dict(torch.load(resnet_weights, map_location=device))

# VGG-19 with BatchNorm (Option A)
try:
    vgg_model = models.vgg19_bn(weights=models.VGG19_BN_Weights.IMAGENET1K_V1)
    print("VGG19_BN: loaded pretrained weights.")
except Exception as e:
    print("VGG19_BN: pretrained weights unavailable, initializing randomly.", e)
    vgg_model = models.vgg19_bn(weights=None)

# Freeze feature extractor (train classifier only) - keeps parity with earlier code where VGG classifier was replaced
for param in vgg_model.features.parameters():
    param.requires_grad = False

# replace final classifier layer
in_features = vgg_model.classifier[6].in_features
vgg_model.classifier[6] = nn.Linear(in_features, num_classes)
vgg_model = vgg_model.to(device)

if os.path.exists(vgg_weights):
    print("Loading VGG checkpoint...")
    vgg_model.load_state_dict(torch.load(vgg_weights, map_location=device))


# ============================================================
#                TRAINING FUNCTION (generic)
# ============================================================
def train_model(model, train_loader, valid_loader, num_epochs=50, save_path=None, patience=7, lr=1e-3):
    criterion = nn.CrossEntropyLoss()
    # only parameters that require grad will be optimized
    optimizer = optim.AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=lr)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', patience=3, factor=0.5, verbose=True)

    best_val_loss = float('inf')
    epochs_no_improve = 0

    history = {"train_loss":[], "val_loss":[], "train_acc":[], "val_acc":[]}

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for imgs, labels in train_loader:
            imgs, labels = imgs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(imgs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, preds = outputs.max(1)
            total += labels.size(0)
            correct += preds.eq(labels).sum().item()

        train_loss = running_loss / len(train_loader)
        train_acc = 100.0 * correct / total

        # validation
        model.eval()
        val_loss = 0.0
        correct = 0
        total = 0
        with torch.no_grad():
            for imgs, labels in valid_loader:
                imgs, labels = imgs.to(device), labels.to(device)
                outputs = model(imgs)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                _, preds = outputs.max(1)
                total += labels.size(0)
                correct += preds.eq(labels).sum().item()

        val_loss = val_loss / len(valid_loader)
        val_acc = 100.0 * correct / total

        scheduler.step(val_loss)

        history["train_loss"].append(train_loss)
        history["train_acc"].append(train_acc)
        history["val_loss"].append(val_loss)
        history["val_acc"].append(val_acc)

        print(f"[{model.__class__.__name__}] Epoch {epoch+1}/{num_epochs} | Train Acc {train_acc:.2f}% | Val Acc {val_acc:.2f}% | Val Loss {val_loss:.4f}")

        # checkpoint
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            epochs_no_improve = 0
            if save_path:
                torch.save(model.state_dict(), save_path)
                print(f"Saved best model to {save_path}")
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= patience:
                print(f"Early stopping after {epoch+1} epochs.")
                break

    return history


# ============================================================
#                TRAIN MODELS
# ============================================================
# NOTE: you can adjust num_epochs and lr per model if you want.
print("\n=== Training CNN ===")
history_cnn = train_model(cnn_model, train_loader, valid_loader, num_epochs=50, save_path=cnn_weights, patience=7, lr=1e-3)

print("\n=== Training ResNet50 ===")
# allow training all params for resnet (we didn't freeze)
history_resnet = train_model(resnet_model, train_loader, valid_loader, num_epochs=50, save_path=resnet_weights, patience=7, lr=1e-4)

print("\n=== Training VGG19_BN (classifier only) ===")
# classifier only (we froze features earlier)
history_vgg = train_model(vgg_model, train_loader, valid_loader, num_epochs=50, save_path=vgg_weights, patience=7, lr=1e-3)


# ============================================================
#                PLOT TRAINING HISTORY
# ============================================================
def plot_history_multi(histories, labels):
    # histories: list of dicts
    plt.figure(figsize=(14,6))
    # loss
    plt.subplot(1,2,1)
    for h,lab in zip(histories, labels):
        plt.plot(h['val_loss'], label=f"{lab} val loss")
    plt.title("Validation Loss Comparison")
    plt.legend()

    # accuracy
    plt.subplot(1,2,2)
    for h,lab in zip(histories, labels):
        plt.plot(h['val_acc'], label=f"{lab} val acc")
    plt.title("Validation Accuracy Comparison")
    plt.legend()
    plt.show()

plot_history_multi([history_cnn, history_resnet, history_vgg], ["CNN","ResNet50","VGG19_BN"])


# ============================================================
#                EVALUATION ON TEST SET
# ============================================================
def load_model_for_eval(path, model_type):
    if model_type == "cnn":
        m = ImprovedCNN(num_classes)
    elif model_type == "resnet":
        try:
            m = models.resnet50(weights=None)
        except:
            m = models.resnet50(weights=None)
        m.fc = nn.Linear(m.fc.in_features, num_classes)
    elif model_type == "vgg":
        try:
            m = models.vgg19_bn(weights=None)
        except:
            m = models.vgg19_bn(weights=None)
        m.classifier[6] = nn.Linear(m.classifier[6].in_features, num_classes)
    else:
        raise ValueError("Unknown model_type")
    checkpoint = torch.load(path, map_location=device)
    m.load_state_dict(checkpoint)
    m = m.to(device)
    m.eval()
    return m

def evaluate_on_loader(model, loader):
    all_preds = []
    all_labels = []
    criterion = nn.CrossEntropyLoss()
    total_loss = 0.0
    with torch.no_grad():
        for imgs, labels in loader:
            imgs = imgs.to(device)
            labels = labels.to(device)
            outputs = model(imgs)
            loss = criterion(outputs, labels)
            total_loss += loss.item()
            _, preds = outputs.max(1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    avg_loss = total_loss / len(loader)
    acc = accuracy_score(all_labels, all_preds) * 100.0
    return acc, avg_loss, np.array(all_labels), np.array(all_preds)


print("\n=== Evaluating models on TEST set ===")
results = []
models_info = [
    ("CNN", cnn_weights, "cnn"),
    ("ResNet50", resnet_weights, "resnet"),
    ("VGG19_BN", vgg_weights, "vgg")
]

for name, path, mtype in models_info:
    if not os.path.exists(path):
        print(f"Skipping {name} evaluation - checkpoint not found at {path}")
        results.append({"model": name, "test_acc": None, "test_loss": None})
        continue

    m = load_model_for_eval(path, mtype)
    acc, loss, y_true, y_pred = evaluate_on_loader(m, test_loader)
    print(f"{name} | Test Acc: {acc:.2f}% | Test Loss: {loss:.4f}")
    # confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
    plt.figure(figsize=(10,10))
    disp.plot(cmap=plt.cm.Blues, xticks_rotation=90)
    plt.title(f"{name} Confusion Matrix (Test)")
    plt.show()

    results.append({"model": name, "test_acc": acc, "test_loss": loss,
                    "y_true": y_true, "y_pred": y_pred})

# Summary DataFrame
summary_rows = []
for r in results:
    summary_rows.append({
        "Model": r["model"],
        "Test Accuracy (%)": None if r["test_acc"] is None else round(r["test_acc"], 3),
        "Test Loss": None if r["test_loss"] is None else round(r["test_loss"], 4)
    })
summary_df = pd.DataFrame(summary_rows)
print("\n=== Performance Summary ===")
print(summary_df)


# ============================================================
#                Ensemble Voting (simple)
# ============================================================
# If all three predictions are available, we can make a simple ensemble by majority vote.
if all(r.get("y_pred") is not None for r in results):
    print("\n=== Simple Majority-Vote Ensemble on Test Set ===")
    preds_stack = np.vstack([r["y_pred"] for r in results])  # shape (3, N)
    # majority vote along axis 0
    from scipy.stats import mode
    ens_pred, _ = mode(preds_stack, axis=0)
    ens_pred = ens_pred.ravel()
    y_true = results[0]["y_true"]
    ens_acc = accuracy_score(y_true, ens_pred) * 100.0
    print(f"Ensemble Test Accuracy: {ens_acc:.2f}%")
else:
    print("\nEnsemble skipped because one or more model predictions are missing.")


# ============================================================
#                PRINT DETAILED TABLE
# ============================================================
print("\nDetailed per-model results (if available):")
for r in results:
    print("-" * 60)
    print(f"Model: {r['model']}")
    if r.get("test_acc") is None:
        print("No checkpoint / results")
        continue
    print(f"Test Accuracy: {r['test_acc']:.3f}%")
    print(f"Test Loss: {r['test_loss']:.4f}")
    # Optionally print top-5 confusion rows or per-class accuracy
    per_class_acc = []
    cm = confusion_matrix(r["y_true"], r["y_pred"])
    class_acc = cm.diagonal() / cm.sum(axis=1)
    # handle division by zero in case a class has zero samples
    class_acc = np.nan_to_num(class_acc)
    for cls, acc_cls in zip(class_names, class_acc):
        per_class_acc.append((cls, round(acc_cls * 100, 2)))
    # print first 10 classes for brevity
    print("Per-class accuracy (first 10 shown):")
    for cls, acc_cls in per_class_acc[:10]:
        print(f"  {cls}: {acc_cls}%")
```
